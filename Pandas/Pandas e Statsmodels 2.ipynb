## Estatísticas Móveis

Quando trabalhamos com dados temporais, é frequente o cáculo de estatísticas sobre janelas móveis. O pandas também nos permite fazer isso com facilidade:

```python
ts = pd.Series(randn(1000), index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot(style='k--');
pd.rolling_mean(ts, 60).plot(style='b');
```

A soma rolante é apenas uma das estatísticas rolantes disponíveis:

```python
[s for s in dir(pd) if s.startswith('rolling')]
```

Isto também funciona para Dataframes, com a mesma simplicidade.

```python
df = pd.DataFrame(randn(1000, 4), index=ts.index,columns=['A', 'B', 'C', 'D'])
```

```python
df = df.cumsum()
pd.rolling_sum(df, 60).plot(subplots=True,figsize=(10,10));
```

Você tabém pode criar suas próprias funções e usar como uma estatística móvel:

```python
mad = lambda x: np.fabs(x - x.mean()).mean()
pd.rolling_apply(ts, 60, mad).plot(style='k');
```

### Estatísticas Móveis Bi-variadas

As estatísticas móveis bi-variadas podem ser calculadas *entre duas séries*, retornando uma série; *Entre um DF e uma série*, retornando um Dataframe; e *entre dois dataframes*, retornando um dataframe da estatística entre as colunas correspondentes.

```python
df2 = df[:20]
pd.rolling_corr(df2, df2['B'], window=5)
```

### Correlaçoes móveis pareadas

Em finanças é costumeiro calcular matrizes de correlações a partir de uma coleção de séries temporais. Mas calcular estas matrizes de maneira móvel, pode ser bem complicado. O pandas facilita tal operação, retornando um **Painel**(dataframe de 3 dimensões) onde os ítems  são as datas:

```python
correls = pd.rolling_corr_pairwise(df, 50)
correls[df.index[-50]]
```

Deste painel pode-se extrair as séries de correlação entre qualquer par de variáveis, facilmente:

```python
correls.ix[:, 'A', 'C'].plot();
```
