## Arquivos de Texto e CSVs

As principais funções para leitura de arquivos de texto são *read_csv* e *read_table*. Ambas usam o mesmo parser para converter inteligentemente dados tabulados para um Dataframe.

```python
df = pd.read_csv('Quantidade_Processos-Comarca-Anos-JEC-2005_a_2000.csv',sep=',',header=0, parse_dates=True)
```

```python
df[0:5]
```

A Primeira coluna do dataframe, *comarca* na verdade é um descritor para os dados da outras linhas, logo fica melhor como indices do dataframe.

```python
df = pd.read_csv('Quantidade_Processos-Comarca-Anos-JEC-2005_a_2000.csv',sep=',',header=0,index_col=0, parse_dates=True)
```

```python
df[0:5]
```

```python
df.describe()
```

```python
df.xs('Nova Iguaçu')
```

## Salvando Dados

Séries e dataframes possuem métodos `to_csv('dados.csv')` e  `to_html()` para gerar exportações dos dados. Arquivos Excel também podem ser fácilmente gerados com `ExcelWriter`

```python
df.to_excel('dados.xlsx', sheet_name='sheet1')
```

## Visualizando dados

```python
df.T['Nova Iguaçu'].plot()
```

Veja como neste gráfico, os anos aparecem não em ordem cronológica mas na ordem em que estão posicionados no dataframe. Para corrigir isso precisamos converter os títulos das colunas para objetos do tipo datetime.

```python
df.columns = pd.to_datetime(df.columns)
```

```python
df.T['Nova Iguaçu'].plot()
```

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(df,alpha=0.2, figsize=(10, 10), diagonal='kde')
```

```python
df[:10].T.plot(logy=True,legend=False, figsize=(10,10))
```
